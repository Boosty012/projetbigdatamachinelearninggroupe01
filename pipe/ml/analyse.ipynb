```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Chargement des données
df = pd.read_csv("../data/data_from_elasticsearch.csv")
```

```python
# Nettoyage des données
print("\nSuppression des doublons...")
df.drop_duplicates(inplace=True)

print("Suppression des lignes avec valeurs manquantes...")
df.dropna(inplace=True)
```

```python
# Enregistrement des données nettoyées
df.to_csv("../data/cleaned_data.csv", index=False)
```

```python
print(df.head())
```

```python
# Distribution des valeurs
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['close'], kde=True)
plt.title("Distribution des prix de clôture")

plt.subplot(1, 2, 2)
sns.histplot(df['volume'], kde=True)
plt.title("Distribution du volume")

plt.tight_layout()
plt.show()
```

```python
# Corrélation entre les variables
print("\nCorrélation entre les variables :")
print(df.corr(numeric_only=True))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Matrice de corrélation")
plt.show()
```
